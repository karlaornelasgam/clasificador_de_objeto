```python
import cv2 as cv

# Carga el archivo cascade.xml
objeto = cv.CascadeClassifier('/Users/karla/Documents/inteligencia_artificial/tareas/pruebas5/classifier/cascade.xml')
#C:\Users\karla\Documents\inteligencia_artificial\tareas\dataset2\classifier
if objeto.empty():
    print("Error al cargar el archivo cascade.xml")
    exit()

# Inicializa la cámara
cap = cv.VideoCapture(0)

while True:
    # Captura el fotograma actual
    ret, frame = cap.read()
    if not ret:
        break

    # Convierte la imagen a escala de grises
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    # Detecta objetos utilizando el archivo cascade.xml
    objetos = objeto.detectMultiScale(gray, 1.3, 5)

    # Dibuja un rectángulo azul alrededor de cada objeto detectado
    for (x, y, w, h) in objetos:
        cv.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Muestra el resultado en una ventana
    cv.imshow('Detección del objeto', frame)

    # Rompe el bucle si se presiona la tecla Esc
    if cv.waitKey(1) == 27:
        break

# Libera los recursos y cierra las ventanas
cap.release()
cv.destroyAllWindows()
```
